```python
import pandas as pd
```

```python
df_weather = pd.read_csv("cleaned_combined_texas_weather_2022_2024.csv", index_col=0, parse_dates=True)
df_cloud = pd.read_csv("synced_texas_cloud_coverage_2022_2024.csv", index_col=0, parse_dates=True)
```

```python
print("Missing Values: {}".format(pd.isna(df_weather.index).sum()))
df_weather.describe()
```

```python
print("Missing Values: {}".format(pd.isna(df_cloud.index).sum()))
df_cloud.describe()
```

```python
df_weather = df_weather.drop(columns=['City']).groupby(df_weather.index).mean()
df_weather
```

```python
df_cloud = df_cloud.mean(axis=1).to_frame(name='Cloud Coverage')
df_cloud
```
